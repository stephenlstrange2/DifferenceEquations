```python
import numpy as np  #necesitamos numpy
import matplotlib.pyplot as plt
```

```python
l = 1000
```

```python
x = np.zeros(l)
y = np.zeros(l)
n = 0
w = 1.6
cosw = np.cos(w)
print("cosw =" ,cosw)
x[0] = 1
y[0] = x[0]
```

```python
def seno(cosw,yn1,yn2,xn,xn1):
    return (2*cosw*yn1)-(yn2)+(xn)-(cosw*xn1)
y[1] = seno(cosw,y[0],0,x[1],x[0])
for i in range(2,l): #argumento range te manda desde 0 hasta l - 1
    y[i] = seno(cosw,y[i-1],y[i-2],x[i],x[i-1])
```

```python
# Data for plotting
t = np.arange(0.0, , 0.01)
# s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

fig.savefig("test.png")
plt.show()
```
